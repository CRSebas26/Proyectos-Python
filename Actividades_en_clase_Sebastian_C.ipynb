```python
import pandas as pd

# Create a DataFrame with fictitious data
datos = {
    'Fecha': pd.date_range(start='2024-01-01', periods=50, freq='D'),
    'Producto': ['Producto_A', 'Producto_B', 'Producto_C', 'Producto_D', 'Producto_E'] * 10,
    'Cantidad': [5, 10, 15, 20, 25] * 10,
    'Precio': [100, 150, 200, 250, 300] * 10
}
df = pd.DataFrame(datos)
df
```

```python
import pandas as pd
import numpy as np

datos = {
    'Nombre': ['Ana', 'Luis', 'Juan', 'Maria', np.nan, 'Pedro', 'Carla'],
    'Edad': [23, np.nan, 35, 29, 42, np.nan, 30],
    'Salario': [50000, 55000, np.nan, 62000, 58000, 60000, np.nan]
}
df = pd.DataFrame(datos)
df
```

```python
nulos_por_columna = df.isnull().sum()
print("Valores nulos por columna:")
print(nulos_por_columna)

total_nulos = df.isnull().sum().sum()
print(f"\nTotal de valores nulos en el DataFrame: {total_nulos}")
```

```python
df_sin_nulos_filas = df.dropna()
print("\nDataFrame después de eliminar filas con valores nulos:")
print(df_sin_nulos_filas)
```

```python
df_rellenado_especifico = df.copy() # Create a copy to avoid modifying the original DataFrame directly
df_rellenado_especifico['Nombre'] = df_rellenado_especifico['Nombre'].fillna('Desconocido')
df_rellenado_especifico['Edad'] = df_rellenado_especifico['Edad'].fillna(0)
df_rellenado_especifico['Salario'] = df_rellenado_especifico['Salario'].fillna(0)

print("\nDataFrame después de rellenar valores nulos con un valor específico:")
print(df_rellenado_especifico)
```

```python
df_rellenado_promedio = df.copy() # Create another copy for this operation

# Calculate the mean for numeric columns with NaNs
mean_edad = df_rellenado_promedio['Edad'].mean()
mean_salario = df_rellenado_promedio['Salario'].mean()

# Fill NaNs with their respective means
df_rellenado_promedio['Edad'] = df_rellenado_promedio['Edad'].fillna(mean_edad)
df_rellenado_promedio['Salario'] = df_rellenado_promedio['Salario'].fillna(mean_salario)

print("\nDataFrame después de rellenar valores nulos con el promedio de la columna:")
print(df_rellenado_promedio)
```
